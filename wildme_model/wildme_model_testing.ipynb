```python
import torch
import lightnet as ln

params = ln.engine.HyperParameters(
    # Network
    class_label_map=['grouper_nassau'],
    input_dimension=(416, 416),
    batch_size=64,
    mini_batch_size=8,
    max_batches=60000,

    # Dataset
    _train_set='/data/db/_ibsdb/_ibeis_cache/training/lightnet/lightnet-training-grouper_nassau-afae6bc69d341095/data/train.pkl',
    _valid_set=None,
    _test_set='/data/db/_ibsdb/_ibeis_cache/training/lightnet/lightnet-training-grouper_nassau-afae6bc69d341095/data/test.pkl',
    _filter_anno='ignore',

    # Data Augmentation
    jitter=.3,
    flip=.5,
    hue=.1,
    saturation=1.5,
    value=1.5,
)

model = ln.models.YoloV4(
    len(params.class_label_map)
)
```

```python
def init_weights(m):
    if isinstance(m, torch.nn.Conv2d):
        torch.nn.init.kaiming_normal_(m.weight)
```

```python
model.apply(init_weights)

weights_file = "wildme_grouper_weights" 
model.load(weights_file)

model.eval()
```

```python
import cv2
import numpy as np
```

```python
image_path = "3_8_2024_0U1A0590_04861.png" 
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
resized_image = cv2.resize(image, params.input_dimension)
resized_image = resized_image.astype(np.float32) / 255.0
resized_image = np.transpose(resized_image, (2, 0, 1)) 
resized_image = np.expand_dims(resized_image, axis=0)
```

```python
input_tensor = torch.from_numpy(resized_image)
```

```python
with torch.no_grad():
    detections = model(input_tensor)
```
