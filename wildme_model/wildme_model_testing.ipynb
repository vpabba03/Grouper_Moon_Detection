```python
params = ln.engine.HyperParameters(
    # Network
    class_label_map=['grouper_nassau'],
    input_dimension=(416, 416),
    # input_dimension=(1000, 2000),
    batch_size=64,
    mini_batch_size=8,
    max_batches=60000,

    # Dataset
    _train_set='/data/db/_ibsdb/_ibeis_cache/training/lightnet/lightnet-training-grouper_nassau-afae6bc69d341095/data/train.pkl',
    _valid_set=None,
    _test_set='/data/db/_ibsdb/_ibeis_cache/training/lightnet/lightnet-training-grouper_nassau-afae6bc69d341095/data/test.pkl',
    _filter_anno='ignore',

    # Data Augmentation
    jitter=.3,
    flip=.5,
    hue=.1,
    saturation=1.5,
    value=1.5,
)

## load model params

model = ln.models.Yolo(
    len(params.class_label_map)
).double()
```

```python
from collections import OrderedDict

## using cpu, no available gpu
device='cpu'
weights_file = "wildme_grouper_weights"


## load in weights, pulling network from the dict
network_weights = torch.load(weights_file, map_location=device)['network']

## delete 'loss.seen' from the weights
if isinstance(network_weights, OrderedDict):
    if 'loss.seen' in network_weights:
        del network_weights['loss.seen']
        print('deleted')
else:
    print("The loaded weights are not in an OrderedDict format.")

try:
    # model.load_state_dict(torch.load(weights_file, map_location=device)['network'])
    model.load_state_dict(network_weights)
    print("Weights loaded successfully")
except Exception as e:
    print(f"Error loading weights: {e}")
```

```python
import cv2
import numpy as np
```

```python
## loading in a frame from a video

image_path = "test_frames/video100201.jpeg" 
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
```

```python
image.shape
```

```python
## preprocessing
image = cv2.resize(image, params.input_dimension)
resized_image = image.astype(np.float64)
resized_image = np.transpose(resized_image, (2, 0, 1)) 
resized_image = np.expand_dims(resized_image, axis=0)
resized_image = np.vstack(resized_image).astype(np.float64)
```

```python
# temp_image = np.transpose(image, (2, 0, 1)) 
```

```python
## convert to input tensor
input_tensor = torch.as_tensor(np.array(resized_image).astype('float')).unsqueeze(0).type(torch.DoubleTensor)
```

```python
input_tensor.shape
```

```python
## run model with the input tensor
with torch.no_grad():
    detections = model(input_tensor)
```

```python
detections
```

```python
detections.shape
```

```python
## Convert Detections into boxes
```

```python
## postprocess found in lightnetv1.0.0 documentation
postprocess = ln.data.transform.Compose([
    ln.data.transform.GetBoundingBoxes(len(params.class_label_map), model.anchors, 0.1), ## adjustable confidence level
    ln.data.transform.NonMaxSuppression(0.5),
    ln.data.transform.TensorToBrambox(params.input_dimension, params.class_label_map)
])
```

```python
## run postprocess on detections
bboxes = postprocess(detections)
bboxes[0:5]
```

```python
## Print boxes on top off an image
```

```python
## draw out boxes iteratively using .rectangle
for idx, row in bboxes.iterrows():
    try:
        x_top_left = int(float(row.iloc[3])) 
        y_top_left = int(float(row.iloc[4]))  
        width = float(row.iloc[5])
        height = float(row.iloc[6])
        
        right_x = int(x_top_left + width)  
        bottom_y = int(y_top_left + height) 
        
        cv2.rectangle(image, (x_top_left, y_top_left), (right_x, bottom_y), color=(255, 0, 0), thickness=1)
    except (ValueError, IndexError) as e:
        print(f"Skipping invalid box: {row} ({e})")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

## show the boxes
cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
