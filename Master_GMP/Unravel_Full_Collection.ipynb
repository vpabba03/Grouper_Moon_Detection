```python
dfs = []

for col in mediaAsset_cols:
    i = col.split('.')[-1][-1]
    columns_to_keep = [col] + [f'keywords{i}'] + [col for col in df.columns if col not in mediaAsset_cols and col not in columns_to_rename]
    df_i = df[columns_to_keep].copy()
    df_i.rename(columns={col: 'mediaAsset', f'keywords{i}': 'keywords'}, inplace=True)
    df_i.dropna(subset=['mediaAsset'], inplace=True)
    #df_i.set_index('mediaAsset', inplace=True)
    dfs.append(df_i)
```

```python
filtered_df = dfs[1][~dfs[1]['mediaAsset'].isin(dfs[0]['mediaAsset'])]
```

```python
#dfs[0].set_index('mediaAsset', inplace=True)
#filtered_df.set_index('mediaAsset', inplace=True)
```

```python
dfs[0] = dfs[0].loc[:,~dfs[0].columns.duplicated()].copy()
dfs[0].head()
```

```python
filtered_df = filtered_df.loc[:,~filtered_df.columns.duplicated()].copy()
filtered_df.head()
```

```python
con_df = pd.concat([filtered_df, dfs[0]], ignore_index=True)
#con_df.set_index('mediaAsset', inplace=True)
con_df.head()
```
