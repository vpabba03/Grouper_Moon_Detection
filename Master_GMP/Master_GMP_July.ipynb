```python
# DataBulkImport, Submitter
def add_date_submitter(file, date, submitter):
    bt_file = pd.read_excel(f'./Bryant.Tran/{file}')
    split_values = bt_file['Encounter.mediaAsset0'].to_list()
    result = df[df['mediaAsset'].isin(split_values)]
    template.loc[template['ImageID'].isin(result['mediaAsset']), 'DateBulkImport'] = date
    template.loc[template['ImageID'].isin(result['mediaAsset']), 'SubmitterName'] = submitter
    return template[~template['DateBulkImport'].isna()].shape
```

```python
# Year, Month, Day, Hour, Minutes
```

```python
template['Date'] = df['indexID'].str.split("_")
template['Date'] = template['Date'].apply(lambda x: x[-4:] if isinstance(x, list) else [])
template['Year'] = template['Date'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
template['Month'] = template['Date'].apply(lambda x: x[1] if isinstance(x, list) and len(x) > 0 else None)
template['Day'] = template['Date'].apply(lambda x: x[2] if isinstance(x, list) and len(x) > 0 else None)
template['Hour'] = df['Encounter.hour']
template['Minutes'] = df['Encounter.minutes']

template.drop(columns=['Date'], inplace=True)
```

```python
# LocationDescription, Country, LocationID, Genus, Species
```

```python
template['LocationDescription'] = np.where(df['indexID'].str.upper().str.contains('BRAC', na=False), 
                                           'East End Cayman Brac FSA', 
                                           "West End Little Cayman FSA")

template['Country'] = 'Cayman Islands'

template['LocationID'] = np.where(df['indexID'].str.upper().str.contains('BRAC', na=False), 
                                  'East End Cayman Brac FSA', 
                                  "West End Little Cayman FSA")

template['Genus'] = df['Encounter.genus']
template['Species'] = df['Encounter.specificEpithet']
```
