```python
def same_as_files(df):
    same_as = df.copy()
    same_as = same_as.iloc[:, 1:]
    df.drop(columns=['same_as_other_box'], inplace=True)
    same_as.rename(columns={'same_as_other_box': 'ImageID'}, inplace=True)
    
    df = df.drop_duplicates(subset='ImageID')
    same_as = same_as.drop_duplicates(subset='ImageID')

    same_as.columns = [f"{col}_dup" if col != 'ImageID' else col for col in same_as.columns]
    concatenated_df = pd.concat([df.set_index('ImageID'), same_as.set_index('ImageID')], axis=1).reset_index()

    columns_to_drop = [col for col in concatenated_df.columns if col.endswith('_dup')]
    concatenated_df.drop(columns=columns_to_drop, inplace=True)
    
    return concatenated_df
```

```python
feb_2020 = pd.read_csv('./Uploaded_Data_Files/all_bboxes_extracted_02_22_2020.csv')
jun_2019 = pd.read_csv('./Uploaded_Data_Files/all_bboxes_extracted_06_03_2020.csv')
oct_2019 = pd.read_csv('./Uploaded_Data_Files/all_boxes_extracted_10_15_2019.csv')
sep_2020 = pd.read_csv('./Uploaded_Data_Files/all_boxes_extracted_2020_09_29.csv')
```

```python
feb_2020.rename(columns={"bbox_name_(auto)":"ImageID", "parent_folders_(auto)":"Parent_Folder"}, inplace=True)
jun_2019.rename(columns={"bbox_name_(auto)":"ImageID", "parent_folders_(auto)":"Parent_Folder"}, inplace=True)
oct_2019.rename(columns={"bbox_name_(auto)":"ImageID", "parent_folders_(auto)":"Parent_Folder"}, inplace=True)
sep_2020.rename(columns={"bbox_name_(auto)":"ImageID", "parent_folders_(auto)":"Parent_Folder"}, inplace=True)
```

```python
feb_2020.columns
```

```python
new_df_1 = same_as_files(sep_2020)
new_df_2 = same_as_files(jun_2019)
new_df_3 = same_as_files(oct_2019)
new_df_4 = same_as_files(feb_2020)
```

```python
combined_df = pd.concat([new_df_1, new_df_2, new_df_3, new_df_4], ignore_index=True)
```

```python
combined_df = combined_df[['ImageID', 'Parent_Folder']]
combined_df.head()
```
