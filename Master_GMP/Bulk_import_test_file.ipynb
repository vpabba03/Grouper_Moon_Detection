```python
import pandas as pd
import numpy as np
```

```python
all_boxes = pd.read_csv("Master_GMP.csv")
boxes_uploaded = pd.read_csv("collection_of_uploaded_bboxes_full.csv")
```

```python
non_matching_in_all_boxes= all_boxes[~all_boxes['ImageID'].isin(boxes_uploaded['mediaAsset'])]
non_matching_in_all_boxes.head()
```

```python
index_counts = non_matching_in_all_boxes['IndexID'].value_counts()
index_counts
```

```python
boxes_mult_fish = non_matching_in_all_boxes[non_matching_in_all_boxes['IndexID'].isin(index_counts[index_counts > 1].index)]
boxes_mult_fish[boxes_mult_fish['IndexID'] == 'GrouperMoon_2015_3v_F6_1pm5pm_014_2015_0_0_1']
```

```python
boxes_mult_fish = boxes_mult_fish[['IndexID','Encounter.occurrenceID', 'Imported Image ID', 'Year', 'Month', 'Day', 'Hour', 'Minutes', 'DecimalLatitude', 'DecimalLongitude', 'LocationDescription', 'Country', 'LocationID',
               'Genus', 'Species', 'IndividualIDGrouperSpotter', 'IndividualNickname', 'Sex', 'VideoComments', 'ImageComments', 'Encounter.submitter.emailAddress', 'PhotographerFullName', 'PhotographerEmail',
               'Fish_Length', 'Distinguishing Features', 'LivingStatus', 'Lifestage', 'FishDepth', 'Behavior', 'TissueSampleID', 'Encounter.mediaAsset0.keywords', 'Encounter.mediaAsset1.keywords']]
```

```python
seen = {}
counter = 0

# Iterate through each row
for index, row in boxes_mult_fish.iterrows():
    indexID = row["IndexID"]
    imported_image_id = row["Imported Image ID"]
    
    if indexID in seen:
        # Increment the count of images for this IndexID
        seen[indexID]['count'] += 1
        # Create the mediaAsset column name dynamically (mediaAsset1, mediaAsset2, etc.)
        media_asset_column = f"mediaAsset{seen[indexID]['count']}"
        
        # If the column doesn't exist yet, create it
        if media_asset_column not in boxes_mult_fish.columns:
            boxes_mult_fish[media_asset_column] = None  # Initialize the column with None values
        
        # Assign the value to the current row in the dynamically created column
        boxes_mult_fish.loc[index, media_asset_column] = imported_image_id
        counter += 1
        print(index, imported_image_id, counter)
    else:
        # Mark this IndexID as seen, storing the current index and initialize count to 1
        seen[indexID] = {'index': index, 'count': 1}
```

```python
boxes_mult_fish.drop(columns=['IndexID'], inplace=True)
```

```python
boxes_mult_fish.insert(2, "Encounter.mediaAsset1", np.nan)
boxes_mult_fish.insert(21, "Encounter.state", np.nan)
boxes_mult_fish.insert(25, "Encounter.measurement2", np.nan)
```

```python
boxes_mult_fish = boxes_mult_fish[~boxes_mult_fish['Encounter.mediaAsset1'].isna()]
boxes_mult_fish[boxes_mult_fish['mediaAsset2'] != boxes_mult_fish['Imported Image ID']]
```
